```python
train = pd.read_csv('/home/kesci/input/a106803/train.csv')
test=pd.read_csv('/home/kesci/input/a106803/test.csv')
submit=pd.read_csv('/home/kesci/input/a106803/submit.csv')
```

```python
train.head()
```

```python
train_eda=train.drop(['id','label'],axis=1)
test_eda=test.drop('id',axis=1)
```

```python
#相关性分析
plt.figure(figsize=(18, 12),)
sns.heatmap(train_eda, cmap='RdBu_r', annot=True, fmt='.2f',annot_kws={'fontsize':14})
plt.tick_params(axis='x',which='major',labelsize=16,labelrotation=90)
plt.tick_params(axis='y',which='major',labelsize=16)
```

```python
train_describe = train.describe()
np.sum(train_describe.loc['count'] != 500)
sns.distplot(train_describe.loc['count'])
```

```python
train.nunique()
```

```python
#探究各特征在训练集和测试集的分布情况，绿色为测试集，红色为训练集
train_eda["oringin"]="train"
test_eda["oringin"]="test"
data=pd.concat([train_eda,test_eda],axis=0,ignore_index=True)
#View data

# Explore feature distibution 
#fig = plt.figure(figsize=(6, 6))
for column in data.columns[0:-1]:
    g = sns.kdeplot(data[column][(data["oringin"] == "train")], color="Red", shade = True)
    g = sns.kdeplot(data[column][(data["oringin"] == "test")], ax =g, color="Green", shade= True)
    g.set_xlabel(column)
    g.set_ylabel("Frequency")
    g = g.legend(["train","test"])
    plt.show()
```
